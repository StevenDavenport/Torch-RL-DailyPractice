## Question 1: Read the code and state the shapes

```python
obs = torch.randn(2, 4, 3)
W = torch.randn(3, 5)
y = obs @ W
```

What are the shapes of `obs`, `W`, and `y`?
Why does this matrix multiply work even though `obs` has 3 dimensions?


````python
# Question 1: Answer and reasoning
import torch 

obs = torch.randn(2, 4, 3)
W = torch.randn(3, 5)
(obs @ W).shape

# obs shape is (2, 4, 3), W is (3, 5) and y is (2, 4, 5)
# The reason for this is that W transforms the input samples from length-3 vectors to length 5.
# As long as the inner dimentions match this runs. 
# Resulting in the outer dimentions of the inputs remaining (2, 4, etc) but with the transformed inner as 3->5.
````
