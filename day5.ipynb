## Question 1: `(B,)` vs `(B, 1)`

Suppose `x` has shape `(B, T) = (2, 4)`.

You also have:
- `a`: shape `(B,) = (2,)`
- `b`: shape `(B, 1) = (2, 1)`

Will `x * a` work directly?
Will `x * b` work directly?
Explain why by comparing dimensions from the right.


````python
# Question 1: Answer and reasoning
import torch 
import torch.nn.functional as F

x = torch.randn(2,4)
a = torch.randn(2,)
b = torch.randn(2, 1)

# On computing x*a or x*b the dimentions will be compared from right to left. 
# With x*b, the feature or right most dimentions do not match, but one of them has the value of 1, so the this also works. 
# It works because the a will be expanded accross the feature dimention.
# When computing x*a, the first comparison will be between 2 and 1, which works becuase of previously mentioned.
# However, then moving left, x has an extra dimention whereas b does not and this stops execution.
# Braudcasting works only when there is a dimention to expand through.
````

## Question 2: Flattening for cross-entropy

You have:
- `logits`: shape `(B, T, A) = (2, 4, 3)`
- `actions`: shape `(B, T) = (2, 4)`

What shapes should they become before one cross-entropy call?
What does one row of the flattened logits represent?


````python
# Question 2: Answer and reasoning
b,t,a = 2,4,3

logits = torch.randn(b,t,a)
actions = torch.randint(0, 3,(b,t))

logits_flat = logits.reshape(-1, a)
actions_flat = actions.reshape(-1)

loss = F.cross_entropy(logits_flat, actions_flat)

# Before the cross_entropy call, it is required that both logits and actions tensors are flattened accross the batch and time dimentions. (B*T,A)&(B*T).
# Each row of the flattened logits is the 3 actions scores for one specific (batch, time) position.
# The matching entry in the flattened actions tensor is the correct action index for  that same position.
````
