## Question 3: Tiny implementation

You have:
- `logits`: shape `(B, T, A)`
- `actions`: shape `(B, T)`

You want the chosen-action logit at each `(batch, time)` position.

Do not over-polish the code. I want:
- what operation you would use
- what shape `actions` needs before the operation
- what the output shape would be
- what one output element means


````python
# Question 3: Answer and reasoning

# Constructing the dummy logits and actions indicies
b, t, a = 2, 4, 6
logits = torch.randn(b, t, a)
actions = torch.randint(0, a, (b, t))

# Use the actions as indicies the select the corrisponding scores from the logits tensor
# But first we unsqueeze the actions tensor so it has the same number of dimentions as the logits tensor.
# We then squeeze the result of the gather to remove the extra trailing dimention
scores = torch.gather(logits, 2, actions.unsqueeze(-1)).squeeze(-1)

# the shape of the output after the squeeze will be (2, 4)
scores_shape = scores.shape

# one output element means the corrisponding raw score from the tensor to the chosen action tensor index
# i.e. one scalar value (raw score) at each of the (b, t) positions. 
scores[0, 0].item()
````
