## Question 1: Indexing vs slicing

You have `x` with shape `(B, T, D) = (2, 4, 3)`.

What is the shape of `x[:, 1]`?
What is the shape of `x[:, 1:2]`?
Why are they different?


````python
# Question 1: Answer and reasoning
import torch

x = torch.randn(2, 4, 3)
print('x: ', x)
#print('x[:, 1]: ', x[:, 1])
print('x[:, 1]: ', x[:, 1])
#print('x[:, 1:2].shape: ', x[:, 1:2].shape)
print('x[:, 1:2]: ', x[:, 1:2])


# x[:, 1] shape will be (2,3)
# x[:, 1] gathers the 1st index of the second dimention of each of the two batches. 2 batches, therefore 2 vectors, each of length 3. Hence the shape.
# x[:, 1] Because we're accessing the second dimention here it is consumed.
# x[:, 1:2] has a different shape of (2,1,3).
# The result of this is almost identical, the values are the same, however because it is a slice it keeps a middle dimention in the place of the second dimention.
# It is not consumed like the prior example.
````

## Question 2: Flattening time and batch

You have `x` with shape `(B, T, D) = (2, 4, 3)`.

If you do `x.reshape(-1, 3)`, what shape do you get?
What do the rows mean after this reshape?


````python
# Question 2: Answer and reasoning
x = torch.randn(2,4,3)

x.reshape(-1,3).shape

# In this example the reshpape flattens the first two dimentions (batch and time) together therefore the shape becomes (B*T, D) or (8, 3).
# How this actually happens under the hood is: in x there are 2*4*3=24 values. with reshape(-1,3) what we're really asking is not to keep the 3 the same,
# But to do 24/3, resolving the -1 to 8. 
````
