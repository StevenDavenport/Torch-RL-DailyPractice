## Question 1: Read the reshape

```python
x = torch.randn(2, 3, 4)
y = x.reshape(-1, 4)
```

What shape is `y`?
What does one row of `y` represent?
Why does `-1` become that value?


````python
# Question 1: Answer and reasoning

# y will be shape: (6, 4)
# one row of y will represent one sample of length-4 the batch time elements
# -1 works out how many total elements/values there are in the tensor:
    # total = 2*3*4 = 24. It then calculates 24/4 as we requested to keep the second dimention of size 4. 
    # 24/4 = 6. Therefore the first dimention is of size 6, and the second is of size 4.
````

## Question 2: Read the gather

```python
logits = torch.randn(2, 3, 5)
actions = torch.randint(0, 5, (2, 3))
chosen = torch.gather(logits, 2, actions.unsqueeze(-1)).squeeze(-1)
```

What shape is `chosen`?
What does `chosen[1, 2]` mean?
Why do we need `unsqueeze(-1)` before `gather`?


````python
# Question 2: Answer and reasoning

# Chosen is of shape(2, 3).
# Chosen[1, 2] is the second batch element, third time element, and will result in a length 5 vector sample. 
# We need to unsqeeze the actions tensor so that it has the same number of dimentions as logits. as:
    # gather requres the input and the index tensors to have the same number of dimentions.
    # (after we squeeze that extra dimention back out, removing the trailing dimention)
````
