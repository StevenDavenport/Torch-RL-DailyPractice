## Question 1: Scalar masked mean

You have:
- `x`: shape `(B, T) = (2, 3)`
- `mask`: shape `(B, T) = (2, 3)` with 0/1 values

You want one scalar: the mean of all valid entries in `x`.

What should the numerator be?
What should the denominator be?
Why should the denominator come from `mask`, not from `(x * mask) != 0`?


````python
# Question 1: Answer and reasoning
import torch

x = torch.randn(2, 3)
mask = torch.randint(0, 2, (2, 3))

masked_x = mask*x
mask_count = mask.sum().item()

avg = masked_x.sum() / mask_count
avg = (masked_x.sum().item() / mask_count)
avg

# The numerator should be the sum of x*mask.
# The denominator needs to then be sum of the mask which will corrispond with the number of non-zero entries in the mask as their values can only be 0 or 1. 
# The denominator should not be (x * mask) != 0 because there may be 0 values in the input tensor, x.
````

## Question 3: `gather` with `(B, T, A)`

You have:
- `logits`: shape `(B, T, A) = (2, 3, 4)`
- `actions`: shape `(B, T) = (2, 3)`

You want one chosen-action logit for each `(batch, time)` position.

What shape should `actions` have after `unsqueeze` before `gather`?
What dimension should you gather over?
What shape is the result before and after squeezing?


````python
# Question 3: Answer and reasoning

b, t, a = 2, 3, 4
logits = torch.randn(b, t, a)
actions = torch.randint(0, a, (b, t))

scores = torch.gather(logits, 2, actions.unsqueeze(-1)).squeeze(-1)
logits, actions, scores

# After unsqueeze actions should have the shape (b, t, 1) as we add a trailing dimention. 
# We should gather over the third dimention (actions) (index=2).
# the result will be of shape (b, t, 1), before squeezing.
# The result will be of shape (b, t) after squeezing.
````
