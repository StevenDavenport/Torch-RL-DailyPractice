## Question 3: Broadcasting check

You have:
- `x`: shape `(B, T, D) = (2, 4, 3)`
- `mask`: shape `(B, T, 1) = (2, 4, 1)`

Will `x * mask` work directly?
Explain broadcasting from the right.


````python
# Question 3: Answer and reasoning

# Broadcasting works from right to left.
# In the rightmost dimention x has a length of 3, while the mask has a length of 1.
# Braoudcasting will work here because the right most dimention of the mask will be repeated 3 times, matching x.
# In x and mask, both B,T dimentions match with values of 2 and 4. So this computes without any changes.

x = torch.randn(2,4,3)
mask = torch.randn(2,4,1)

(x*mask).shape
````
